```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data = pd.read_excel("Online retail.xlsx")

data.head()
```

```python
data.info()
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
data = data.drop_duplicates()
```

```python
transactions = data.iloc[:,0].astype(str).apply(lambda x: x.split(','))

transactions = transactions.tolist()

transactions[:5]
```

```python
te = TransactionEncoder()

te_array = te.fit(transactions).transform(transactions)

df = pd.DataFrame(te_array, columns=te.columns_)

df.head()
```

```python
frequent_itemsets = apriori(df, min_support=0.02, use_colnames=True)

frequent_itemsets
```

```python
frequent_itemsets.sort_values('support', ascending=False)
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)

rules.head()
```

```python
rules.sort_values('lift', ascending=False)
```

```python
strong_rules = rules[(rules['lift'] > 1) & (rules['confidence'] > 0.5)]

strong_rules
```

```python
for index, row in strong_rules.iterrows():
    print(f"If a customer buys {row['antecedents']} they are likely to buy {row['consequents']}")
```

Interview Questions
1️⃣ What is lift and why is it important in Association rules?

Lift measures the strength of association between two items. It compares the observed frequency of items appearing together with the expected frequency if the items were independent.

Lift(A → B) = Confidence(A → B) / Support(B)

Lift > 1 indicates a positive association between items.

2️⃣ What is support and confidence? How do you calculate them?

Support measures how frequently an itemset appears in the dataset.

Support(A) = Number of transactions containing A / Total transactions

Confidence measures how often item B appears when item A is purchased.

Confidence(A → B) = Support(A ∪ B) / Support(A)

3️⃣ What are some limitations of association rule mining?

It may generate a large number of rules.

High computational cost for large datasets.

Some interesting patterns may be ignored due to low support.

Interpretation of rules can be difficult.
